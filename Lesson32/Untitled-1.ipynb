```python
from time import sleep
```

```python
data = []
for page in range(1,6):
    print(f'Page {page}')
    url = f'https://www.kinopoisk.ru/lists/movies/series-top250/?page={page}'
    r = requests.get(url, headers=headers, cookies=cookies)
    soup = BeautifulSoup(r.text, 'lxml')
    films = soup.findAll('div', class_='styles_root__ti07r')
    for film in films:
        russian_name = film.find('div', class_='base-movie-main-info_mainInfo__ZL_u3').text
        link = 'https://www.kinopoisk.ru' + film.find('a', class_='styles_poster__gJgwz styles_root__wgbNq').get('href')
        try:
            original_name = film.find('span' ,class_='desktop-list-main-info_secondaryTitle__ighTt').text
        except:
            original_name = russian_name
        rate = film.find('span', 'styles_kinopoiskValuePositive__vOb2E styles_kinopoiskValue__9qXjg styles_top250Type__mPloU').text
        
        data.append([link, russian_name, original_name, rate])
    sleep(5)
         
```
